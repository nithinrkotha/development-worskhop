# Challenge: NFL Salaries


Description:

The dataset contains the salaries for the top 100 players for each position for the years 2011-2018.

Challenge:

1) Initial Data Setup

  +  Clean the column names of all columns by converting them to all lower case 

2) What is the average salary for quarterbacks (Columnname : Quarterback) by year?

3) What is the average salary by year for each position, excludes special teamers (Columnname : Special Teamer)?
  +  Hint: Transpose the dataset into "long" format to help with this step.
  
4) Convert the average salary by year for each position to a string prefixed with $ (e.g. 100000 becomes $100000)



```python
#import os
#os.getcwd()
import pandas as pd
salaries_raw_df = pd.read_excel('nfl_salaries.xlsx')
salaries_raw_df.columns = salaries_raw_df.columns.str.lower()
salaries_raw_df.head()
```

```python
avg_quarterback_salaries_df=salaries_raw_df.groupby(['year']).agg({'quarterback':'mean'}).rename(columns={'quarterback':'Avg_Sal'})
avg_quarterback_salaries_df
```

```python
columns_list = salaries_raw_df.columns.to_list()
columns_list.remove('year')
columns_list.remove('special teamer')
columns_list
```

```python
#list comprehension

columns_trial_list = salaries_raw_df.columns.to_list()
unwanted_columns = {'year', 'special teamer'} 
  
columns_trial_list = [ele for ele in columns_trial_list if ele not in unwanted_columns]
columns_trial_list
```

```python
transposed_df = pd.melt(salaries_raw_df, id_vars=['year'], value_vars=columns_list,var_name='Position', value_name='Salary')

transposed_df.head()
```

```python
avg_salaries_df = transposed_df.groupby(['year', 'Position'])['Salary'].mean().to_frame('avg_salaries').reset_index().round(2)

avg_salaries_df['avg_salaries'] = '$' + avg_salaries_df['avg_salaries'].astype(str)
avg_salaries_df.head()
```
